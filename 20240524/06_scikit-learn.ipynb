```python
# 회귀 모델 실습

import numpy as np
import matplotlib.pyplot as plt
r = np.random.RandomState(10)
x = 10 * r.rand(100)
y = 2 * x - 3 * r.rand(100)
plt.scatter(x,y)
```

```python
x.shape
y.shape # 1차원 벡터
```

```python
#  모델 객체를 생성 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
# 모델 훈련
```

```python
#model.fit(x, y) # 에러남. 행렬로 바꿔줘야 함
# type(x) #type(x)

X = x.reshape(100,1)
model.fit(X,y)
```

```python
# 예측
# test 데이터 생성 
# linspace start, end, num 의 매개변수로  시작값 끝값, start, end,의 몇개의 일정한 간격으로 요소를 만들것인지/
# (2, -1), (-1, 50)을  인자로 넣으면 (2, 50)의 형태로 자동으로 변환
x_new = np.linspace(-1, 11, 100)
X_new = x_new.reshape(100,1)
y_new = model.predict(X_new)
```

```python
# 성능 평가 
#  회귀 모델의 경우 RMSE(Root Mean Square Error) 를 사용해 성능을 평가

from sklearn.metrics import mean_squared_error

# 실제 값
#a = y_new.copy() * 10
a= X_new
# 예측값 
b = y_new

def RMSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    rmse = np.sqrt(mse)
    return rmse

error = RMSE(a, b)

print(error)
```

```python
# Estimator : 모델 훈련에서 핵심이 되는 API

from sklearn.datasets import load_wine
data = load_wine()
print(data.data.shape)
print(data.target.shape)
```

```python
X_train = data.data[:142]
X_test = data.data[142:]
print(X_train.shape, X_test.shape)
```

```python
y_train = data.target[:142]
y_test = data.target[142:]
print(y_train.shape, y_test.shape)
```

```python
# 모델 실행 및 예측
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
# 평가
from sklearn.metrics import accuracy_score

print("정답률=", accuracy_score(y_test, y_pred))
```
